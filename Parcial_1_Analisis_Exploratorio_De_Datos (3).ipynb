```python
#A partir de este momento haremos pruebas de significante A/B donde escogimos el Radio del planeta con la masa
# Ho: No hay relación entre el radio y la masa de los planetas
# Ha: Hay relación entre el radio y la masa de los planetas
col1 = dff.pop("Rp")
col2 = dff.pop("Mp")
print(col1)
print(col2)
```

```python
alpha = 0.05
stat, p = ttest_ind(col1, col2)
print('t=%.3f, p=%.3f' % (stat, p))

if p > alpha:
    print('Aceptamos la hipótesis nula.')
else:
    print('Rechazamos la hipótesis nula')
```

```python
#Comenzamos a hacer la prueba ANOVA donde consideramos la Teff, Distance y K_meg
print('Teff media y desviación típica por Distance')
dff.groupby('Distance')['Teff'].agg(['mean', 'std'])
```

```python
print('Teff media y desviación típica por K_mag')
dff.groupby('K_mag')['Teff'].agg(['mean', 'std'])
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
fig = interaction_plot(
    x        = dff.Distance,
    trace    = dff.K_mag,
    response = dff.Teff,
    ax       = ax,
)
```

```python
# Test ANOVA de dos vías (Two-way ANOVA)
# ==============================================================================
pg.anova(
    data     = dff,
    dv       = 'Teff',
    between  = ['Distance', 'K_mag'],
    detailed = True
).round(4)
```
